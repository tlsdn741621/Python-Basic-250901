```python
# Matplotlib(맷플롯립)은
# 파이썬에서 데이터를 시각화하여 그래프로 나타내는 데 가장 널리 사용되는 라이브러리입니다.
# 라인 플롯, 바 차트, 히스토그램, 산점도 등 거의 모든 종류의 그래프를 그릴 수 있으며,
# 데이터 분석 결과를 시각적으로 표현하고 인사이트를 얻는 데 필수적입니다.

# 기본 개념 및 설명
# 1. Matplotlib의 역할 🎨
#
# Matplotlib는 숫자 데이터를 그림으로 바꾸는 다리 역할을 합니다.
# 복잡한 숫자나 표만으로는 파악하기 어려운 데이터의 패턴, 추세, 관계 등을
# 그래프를 통해 한눈에 이해할 수 있게 도와줍니다.
#
# 2. 핵심 구성 요소
#
# Matplotlib 그래프는 보통 두 가지 주요 요소로 구성됩니다.
#
# Figure (피규어): 그래프가 그려지는 전체 캔버스나 종이입니다.
#하나의 Figure 안에는 여러 개의 그래프(Axes)가 포함될 수 있습니다.
#
# Axes (축): 실제 데이터가 그려지는 플롯 공간입니다.
# x축, y축, 눈금, 레이블 등 그래프의 모든 구성 요소를 포함합니다.
# '하나의 그래프'라고 생각하면 쉽습니다.

# 설치
# pip install matplotlib

# 사용법
# import matplotlib.pyplot as plt
```

```python
# 간단한 사용 예시: 꺾은선 그래프 그리기
# 가장 기본적인 꺾은선 그래프(Line Plot)를 그리는 과정은 다음과 같습니다.
#
# 1. 데이터 준비: x축과 y축에 해당하는 데이터를 준비합니다.
# (주로 리스트나 NumPy 배열 사용)
# 2. 플롯 생성: plt.plot() 함수를 사용해 데이터를 그래프로 그립니다.
# 3. 그래프 꾸미기 (선택 사항):
# plt.title(), plt.xlabel(), plt.ylabel() 등으로 제목과 축 이름을 추가합니다.
# 4. 그래프 보여주기: plt.show() 함수를 호출하여 화면에 그래프를 표시합니다.

import matplotlib.pyplot as plt

# 1. 데이터 준비 (x축: 월, y축: 월별 매출)
months = [1, 2, 3, 4, 5, 6]
sales = [150, 220, 280, 250, 310, 350]

# 2. 꺾은선 그래프 생성
plt.plot(months, sales)

# 3. 그래프 꾸미기
plt.title("Monthly Sales Trend") # 제목 추가
plt.xlabel("Month")              # x축 이름 추가
plt.ylabel("Sales (in millions)") # y축 이름 추가

# 4. 그래프를 화면에 표시
plt.show()
```

```python
# 예시 2
# . Matplotlib 주요 그래프 종류 및 사용 예제
#
#
# ① 라인 그래프 (Line Plot)
# 데이터의 추세를 확인하는 데 유용합니다.

import matplotlib.pyplot as plt

# 1. 라인 그래프
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.figure(figsize=(8, 5))

plt.plot(x, y, marker='o', linestyle='-', color='b', label="데이터 변화")

plt.xlabel("시간", fontproperties=font_prop)
plt.ylabel("값", fontproperties=font_prop)
plt.title("라인 그래프 예제", fontproperties=font_prop)
plt.legend(prop=font_prop)
plt.grid(True)
plt.show()

# 주요 옵션
#
# 📌 추가 옵션
#
# linewidth=2 → 선 두께 조절
# alpha=0.8 → 선 투명도 조절
# markerfacecolor='yellow' → 마커 내부 색상
# markeredgecolor='black' → 마커 테두리 색상
# markersize=8 → 마커 크기 조절
# linestyle='--' → 점선 스타일 설정 ('--', '-.', ':')
# xlim([xmin, xmax]) → X축 범위 설정
# ylim([ymin, ymax]) → Y축 범위 설정
# grid(axis='y', linestyle='--', alpha=0.5) → Y축에만 점선 그리드 추가
# plt.fill_between(x, y, color='lightblue', alpha=0.3) → 선 아래 영역 색상 채우기
#
# marker='o' → 데이터 포인트를 원형으로 표시
# linestyle='-' → 선 스타일 설정 (--, -., : 등 가능)
# color='b' → 선 색상 설정 (예: r(빨강), g(초록), b(파랑), k(검정))
```

```python
#예시 4
# ③ 히스토그램 (Histogram)
# 데이터의 분포를 확인하는 데 사용됩니다.
# 히스토그램(Histogram)은
# 연속된 숫자 데이터의 분포를 시각적으로 표현하는 막대그래프의 한 종류입니다.
# 📊 특정 데이터가 어떤 구간에 얼마나 많이 분포하는지(빈도수)를
# 한눈에 파악할 수 있게 해줍니다.

# 예를 들어,
# 우리 반 학생들의 키 데이터를 가지고 있다면,
# "160cm에서 165cm 사이인 학생은 몇 명인가?"와
# 같은 질문에 대한 답을 시각적으로 보여주는 데 매우 유용합니다.

import numpy as np

# 3. 히스토그램
data = np.random.randn(1000)
plt.figure(figsize=(8, 5))

plt.hist(data, bins=20, color="blue", alpha=0.7, edgecolor="black")

plt.xlabel("값 범위", fontproperties=font_prop)
plt.ylabel("빈도수", fontproperties=font_prop)
plt.title("히스토그램 예제", fontproperties=font_prop)
plt.show()

# 주요 옵션
#
# bins=20 → 구간 개수 설정
# edgecolor="black" → 막대 테두리 색상
#
# 📌 추가 옵션
#
# density=True → 확률 밀도 함수로 변환
# cumulative=True → 누적 분포 표시
# log=True → Y축 로그 스케일 적용
# rwidth=0.9 → 막대 간 간격 조절
# plt.axvline(x=np.mean(data), color='r', linestyle='--') → 평균선 추가
# plt.axvline(x=np.median(data), color='g', linestyle='-.') → 중앙값 표시
# plt.xticks(range(-3, 4)) → X축 범주 설정
# plt.grid(axis='y', linestyle='--', alpha=0.5) → Y축에만 점선 그리드 추가
# plt.hist(data, bins=30, histtype='step', color='black') → 윤곽선 스타일 히스토그램
# plt.fill_between(hist_x, hist_y, color='blue', alpha=0.3) → 영역 색상 채우기
```

```python
# 예시 5
# ④ 산점도 (Scatter Plot)
# 두 변수 간의 관계를 시각화할 때 사용됩니다.
#
# 산점도(Scatter Plot)는
# 두 개의 서로 다른 숫자 변수 사이의 관계를
# 시각적으로 보여주는 그래프입니다.
# 📊 각 데이터 포인트를 x축 값과 y축 값에 해당하는 위치에 점으로 찍어 나타냅니다.
#
# 주로 두 변수 간에 어떤 연관성이 있는지,
# 즉 상관관계를 파악하는 데 사용됩니다.
# 예를 들어, '공부 시간'과 '시험 점수'
# 사이의 관계를 알아보고 싶을 때 매우 유용합니다.
#
# ## 산점도의 핵심 구성 요소
# x축 변수: 관계를 파악하고자 하는 두 변수 중 하나입니다. (독립 변수)
#
# y축 변수: 다른 하나의 변수입니다. (종속 변수)
#
# 점 (Point): 각각의 데이터 하나를 의미합니다.
# 한 점은 한 쌍의 (x, y) 값을 가집니다. (예: 어떤 학생의 공부 시간과 시험 점수)
#
# ## 산점도를 통해 알 수 있는 것
# 산점도에 찍힌 점들의 전체적인 패턴을 통해 두 변수 간의 관계를 직관적으로 파악할 수 있습니다.
#
# 양의 상관관계 (Positive Correlation)
# 한 변수가 증가할 때 다른 변수도 함께 증가하는 경향을 보입니다.
# 점들이 오른쪽 위로 향하는 패턴을 보입니다.
# (예: 공부 시간이 늘어날수록 시험 점수도 높아진다.)
#
# 음의 상관관계 (Negative Correlation)
# 한 변수가 증가할 때 다른 변수는 감소하는 경향을 보입니다.
# 점들이 오른쪽 아래로 향하는 패턴을 보입니다.
# (예: 게임 시간이 늘어날수록 시험 점수는 낮아진다.)
#
# 상관관계 없음 (No Correlation)
# 두 변수 사이에 특별한 관계가 보이지 않습니다. 점들이 뚜렷한 패턴 없이 흩어져 있습니다.
# (예: 키와 시험 점수는 별다른 관계가 없다.)

# 4. 산점도
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(8, 5))

plt.scatter(x, y, color="red", alpha=0.7)

plt.xlabel("X 값", fontproperties=font_prop)
plt.ylabel("Y 값", fontproperties=font_prop)
plt.title("산점도 예제", fontproperties=font_prop)
plt.grid(True)
plt.show()

# 주요 옵션
#
# alpha=0.7 → 투명도 조절
# color="red" → 점 색상 지정
#
# 📌 추가 옵션
#
# s=100 → 점 크기 조절
# c=np.random.rand(50) → 점 색상을 값에 따라 다르게 설정
# cmap='viridis' → 컬러맵 설정 ('coolwarm', 'plasma', 'cividis')
# edgecolors='black' → 점 테두리 색상 지정
# linewidths=0.5 → 점 테두리 두께 조절
# alpha=0.6 → 점 투명도 조절
# plt.colorbar() → 컬러 바 추가 (값이 색으로 표현될 경우)
# plt.xscale('log') → X축 로그 스케일 적용
# plt.scatter(x, y, marker='^') → 마커 스타일 변경 ('o', '^', 's', 'd')
# plt.axhline(y=0.5, color='r', linestyle='--') → 특정 y값 기준선 추가
```

```python
# 예시 6
#  원형 그래프 (Pie Chart)
# 비율을 나타낼 때 유용합니다.

# 5. 원형 그래프
labels = ["A", "B", "C", "D"]
sizes = [40, 30, 20, 10]
plt.figure(figsize=(7, 7))

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple'])

plt.title("원형 그래프 예제", fontproperties=font_prop)
plt.show()

# 주요 옵션
#
# autopct='%1.1f%%' → 퍼센트 표시
# startangle=140 → 시작 각도 지정
#
# 📌 추가 옵션
#
# explode=[0, 0.1, 0, 0] → 특정 조각 강조
# shadow=True → 그림자 효과 추가
# colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'] → 색상 지정
# startangle=90 → 시작 각도 변경
# wedgeprops={'edgecolor': 'black', 'linewidth': 1.5} → 테두리 색 및 두께 조절
# plt.pie(sizes, counterclock=False) → 시계 반대 방향으로 그리기
# plt.pie(sizes, autopct=lambda p: '{:.0f}'.format(p * sum(sizes) / 100)) → 실제 값 표시
# plt.legend(loc='upper right') → 범례 위치 설정
# plt.pie(sizes, labels=labels, pctdistance=0.85) → 퍼센트 값 중앙 정렬
# plt.gca().set_aspect('equal') → 정원 형태 유지
```

```python
# 예시 7
# ⑥ 히트맵 (Heatmap)

# 데이터 행렬을 시각화할 때 사용됩니다.
# 히트맵(Heatmap)은
# 데이터 값의 크기를 색상으로 표현하는 시각화 기법입니다.
# 📜 주로 표나 행렬 형태의 데이터에서 값이 높은 부분과
# 낮은 부분을 한눈에 파악할 수 있도록 도와줍니다.
#
# "열(Heat) 지도"라는 이름처럼,
# 값이 높을수록 뜨거운 색(예: 붉은색)으로,
# 값이 낮을수록 차가운 색(예: 푸른색)으로 나타내는 경우가 많습니다.
#
# ## 히트맵의 핵심 구성 요소
# 격자 (Grid): 데이터가 담겨 있는 2차원 공간으로,
# 각 셀(칸)이 하나의 데이터 값을 가집니다. (가로축과 세로축은 변수를 나타냅니다.)
#
# 색상 척도 (Color Scale):
# 데이터 값의 크기를 나타내는 색상의 범례입니다.
# 이 척도를 통해 특정 색상이 어떤 값의 범위를 나타내는지 알 수 있습니다.
#
# 값 (Value):
# 각 셀에 표시될 수 있는 실제 데이터 값입니다. (선택 사항)
#
# ## 히트맵의 주요 용도
# 히트맵은 복잡한 데이터 속에서
# 패턴을 직관적으로 발견하는 데 매우 유용합니다.
#
# 상관관계 분석
# 여러 변수들 사이의 상관관계를 행렬로 만든 후 히트맵으로 시각화하면,
# 어떤 변수들끼리 강한 양 또는 음의 상관관계를 갖는지 쉽게 파악할 수 있습니다.
#
# 웹사이트 사용자 행동 분석
# 웹페이지에서 사용자들이 마우스를 많이 움직이거나
# 클릭하는 영역을 붉은색으로 표시하여,
# 어떤 부분에 관심이 집중되는지 분석할 수 있습니다.
#
# 지리적 데이터 시각화
# 지도 위에 지역별 인구 밀도나
# 질병 발생률 같은 데이터를 색상으로 표현하여 지역적 편차를 쉽게 확인할 수 있습니다.

# 6. 히트맵
import seaborn as sns
data = np.random.rand(5, 5)
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("히트맵 예제", fontproperties=font_prop)
plt.show()


# 주요 옵션
#
# annot=True → 각 셀에 값 표시
# cmap="coolwarm" → 색상 테마 변경
#
# 📌 추가 옵션
#
# linewidths=0.5 → 셀 간 경계선 설정
# linecolor='black' → 경계선 색상 지정
# cbar_kws={'shrink': 0.8} → 컬러바 크기 조절
# annot_kws={'size': 12} → 텍스트 크기 조절
# xticklabels=['A', 'B', 'C', 'D', 'E'] → X축 레이블 변경
# yticklabels=['V1', 'V2', 'V3', 'V4', 'V5'] → Y축 레이블 변경
# mask=np.triu(np.ones_like(data, dtype=bool)) → 상삼각형 마스킹
# center=0 → 중심값 설정 (0을 기준으로 색상 표현)
# cmap='coolwarm' → 색상 테마 변경
# plt.xticks(rotation=45) → X축 레이블 회전
```

```python
# 예시 9
# ② 시계열 데이터 분석 (이동 평균선 추가)
# 이동평균선(Moving Average, MA)은
# 주식 차트 등에서 단기적인 가격 변동(노이즈)을 제거하고
# 전반적인 추세의 방향을 파악하기 위해 사용되는 가장 기본적인 기술적 지표입니다. 📈

# 이름 그대로,
# 일정 기간 동안의 가격을 산술 평균한 값을 선으로 연결한 것입니다.
# "이동(Moving)"이라는 단어가 붙는 이유는
# 새로운 날의 데이터가 추가되면 가장 오래된 날의 데이터는
# 제외되는 방식으로 평균이 계속해서 이동하기 때문입니다.

# 핵심 원리
# 이동평균선은 계산 방식이 매우 간단합니다.
#
# 5일 이동평균선: 최근 5일간의 종가를 모두 더한 후 5로 나눈 값.
#
# 20일 이동평균선: 최근 20일간의 종가를 모두 더한 후 20으로 나눈 값.
#
# 이 과정을 매일 반복하여 계산된 평균값들을
# 선으로 이으면 이동평균선이 만들어집니다.
# 이 선은 매일의 주가 등락보다 훨씬 부드럽게 움직여, 시장의 큰 흐름을 보여줍니다.

# 8. 시계열 데이터 분석
import numpy as np
import pandas as pd

data = np.cumsum(np.random.randn(100)) # 100개의 랜던함 실수
window = 10 # 이동 평균선, 10일의 이동 평균선
rolling_mean = pd.Series(data).rolling(window=window).mean()
# 데이터에 **이동창(rolling window)**을 적용합니다.
# window=10으로 설정했으므로,
# 데이터를 10개씩 묶어 하나의 창으로 만듭니다.
# 첫 번째 창은 0~9번 데이터,
# 두 번째 창은 1~10번 데이터,
# ... 와 같이 한 칸씩 이동하며 창을 구성합니다.

plt.figure(figsize=(10, 5))
plt.plot(data, label="원본 데이터", alpha=0.5)
plt.plot(rolling_mean, label=f"{window}일 이동 평균", color="red")
plt.legend(prop=font_prop)
plt.title("시계열 데이터 분석", fontproperties=font_prop)
plt.show()
```
